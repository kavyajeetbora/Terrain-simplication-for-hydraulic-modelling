## First we have convert Line to polygon to get the width of the road with LineString polygon 

```python
# using buffer to get the width of the road 
buffer_distance=2  #the distance will be in meters for the map 
sf_utm['polygon']=sf_utm.geometry.buffer(buffer_distance)
#line_sf=sf_utm.buffer(degree)
sf_utm.head()
```

## Deining the function to convert the Stringline to Point as per user input data 

```python
#convert the polygon to linestring 
def poly_to_line(polygon):
    x,y=polygon.exterior.coords.xy
    line_string=LineString([Point(x,y) for x,y in zip(x,y) ])
    return line_string
#apply lambada function 
#   poly_to_line(sf_utm['polygon'].iloc[0]) # this was testing code for one polygon only
#merge all the points in a new column and then export to get the points 
```

```python
#function for making Linstring to points 
def mul_point(line,num_spacing=0.5):
    distances = np.arange(0, line.length, num_spacing)
    points = [line.interpolate(distance) for distance in distances]
    multipoint = unary_union(points)  # or new_line = LineString(points)
    return multipoint
#sf_utm=mul_point(sf_utm["geometry"])
```

## calling the mul_point to convert the geometry to points 

```python
#x,y=sf_utm['polygon'].iloc[0].exterior.coords.xy
```

```python
sf_utm['Linestring']=sf_utm['polygon'].apply(lambda x: poly_to_line(x)).to_crs("EPSG:3857")
#converting the center line to points 
sf_utm.points.append(sf_utm['geometry'].apply(lambda x: mul_point(x)).to_crs("EPSG:3857"))
#converting the outer line of polygon to points 
```
