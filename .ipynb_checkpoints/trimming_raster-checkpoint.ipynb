## Reading the tiff file

```python
%%time
filename = "Danmark_Hav_Nakskov.tif"
directory = r"..\Data\NAKSKOV"
file_path = os.path.join(directory,filename)

img_array, src = pytif.read_raster(file_path)
```

```python
print(img_array.min(), img_array.max())
```

## Cleaning the image

```python
%%time
x0,y0 = np.where(np.abs(img_array)>9999)
img_arr = img_array.copy()
img_arr[x0,y0] = 0

print(img_arr.min(), img_arr.max())
```

```python
%%time
pytif.display_image(img_arr)
```

## Trimming the raster

```python
l,b,r,t = src.bounds
l,b,r,t
```

```python
lat1,lat2 = 13500, 18000
long1, long2 = 4500, 10000
urban_img = img_arr[lat1:lat2, long1:long2]


fig = pytif.display_image(urban_img)
fig
```
